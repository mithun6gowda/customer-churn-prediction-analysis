```python
count_col =[]
hist_col = []
for column in df.columns:
    unique_values = df[column].nunique()
    if unique_values <= 20:
        count_col.append(column)
    else:
        hist_col.append(column) 
```

```python
from sklearn.impute import SimpleImputer
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import RandomForestRegressor  
```

```python
def fill_missing_values (df,random_state=42):
    numeric_columns= df.select_dtypes(include=['float64','int64']).columns.tolist()
    categorical_columns=df.select_dtypes(include=['object']).columns.tolist()
    
    
    numeric_imputer = SimpleImputer(strategy='mean')
    df[numeric_columns] = numeric_imputer.fit_transform(df[numeric_columns])
        
    
    for column in categorical_columns:
        if df[column].dtype == 'object':
           encoded_cols = pd.get_dummies(df[column],prefix=column)
           df =pd.concat([df.drop(column ,axis=1),encoded_cols],axis=1)
           
    randomforest_imputer =  IterativeImputer(estimator=RandomForestRegressor(random_state=random_state))
    df = pd.DataFrame(randomforest_imputer.fit_transform(df),columns=df.columns)
    
    return df

df=fill_missing_values(df, random_state=42)
df.head()
    
```

# Model Building

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score , classification_report ,confusion_matrix, ConfusionMatrixDisplay
from sklearn.preprocessing import StandardScaler, OneHotEncoder  
from sklearn.compose import ColumnTransformer  
from sklearn.model_selection import GridSearchCV 
from xgboost import XGBClassifier 
```

```python
X = df.drop(columns=["churn"])
Y = df["churn"]

X_train ,X_test,Y_train ,Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
from imblearn.over_sampling import SMOTE
print('Before upsampling count of label 0 {}'.format(sum(Y_train==0)))
print('Before upsampling count of label 1 {}'.format(sum(Y_train==1)))



sm = SMOTE(sampling_strategy = 1, random_state=1)   
X_train_s, Y_train_s = sm.fit_resample(X_train, Y_train.ravel())
                                         
print('After upsampling count of label 0 {}'.format(sum(Y_train_s==0)))
print('After upsampling count of label 1 {}'.format(sum(Y_train_s==1)))

```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn import feature_selection
from sklearn import model_selection
from sklearn import metrics

from sklearn.linear_model import LogisticRegressionCV, RidgeClassifierCV, SGDClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import AdaBoostClassifier, BaggingClassifier, GradientBoostingClassifier, RandomForestClassifier 
from xgboost import XGBClassifier
```

```python
models=[
    #Ensemble
    AdaBoostClassifier(),
    BaggingClassifier(),
    GradientBoostingClassifier(),
    RandomForestClassifier(),
    
    #Linear Models
    LogisticRegressionCV(),
    RidgeClassifierCV(),
    
    #Nearest Neighbour
    KNeighborsClassifier(),
    
    #XGBoost
    XGBClassifier()
]

metrics_colums =['model_name','test_accuracy','test_precision','test_recall','test_f1']

model_name=[]
test_accuracy =[]
test_precision=[]
test_recall =[]
test_f1=[]
```

```python
scoring = ['accuracy', 'precision', 'recall', 'f1']

results = []

for model in models:
    cv = model_selection.cross_validate(
        model, X, Y,
        cv=5,
        scoring=scoring,
        return_train_score=False
    )
    
    results.append({
        "model": model.__class__.__name__,
        "test_accuracy": cv['test_accuracy'].mean() * 100,
        "test_precision": cv['test_precision'].mean() * 100,
        "test_recall": cv['test_recall'].mean() * 100,
        "test_f1": cv['test_f1'].mean() * 100
    })

# Create DataFrame directly
model_metrics = pd.DataFrame(results)

# Round once at the end (cleaner + avoids repeated ops)
model_metrics = model_metrics.round(3)

# Sort
model_metrics = model_metrics.sort_values(by='test_accuracy', ascending=False)

# Styling
metrics_styled = model_metrics.style.background_gradient(
    subset=['test_accuracy', 'test_f1'],
    cmap='summer'
)

metrics_styled
```

```python
final_model = RandomForestClassifier()
final_model.fit(X_train,Y_train)
train_pred = final_model.predict(X_train)
test_pred = final_model.predict(X_test)

final_model.score(X_test,Y_test)
```

```python
final_model = XGBClassifier()
final_model.fit(X_train,Y_train)
train_pred = final_model.predict(X_train)
test_pred = final_model.predict(X_test)

final_model.score(X_test,Y_test)
```

```python

print("Feature Importance Gain")
feature_important = final_model.get_booster().get_score(importance_type="gain")
keys = list(feature_important.keys())
values = list(feature_important.values())

data = pd.DataFrame(data=values, index=keys, columns=["score"]).sort_values(by = "score", ascending=False)
data.nlargest(20, columns="score").plot(kind='barh', figsize = (15,10))
```
